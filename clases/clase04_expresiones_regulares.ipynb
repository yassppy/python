# Expresión Regular

Es un patrón de búsqueda que le dice a Python qué texto encontrar dentro de una cadena. En vez de buscar una palabra exacta, describes la forma que tiene el texto.

## Módulo re

Python trae incluido el módulo re para trabajar con expresiones regulares

```python
import re
```

### Caracteres básicos

| Patrón | Significa | Ejemplo | Encuentra |
|--------|-----------|---------|-----------|
| `.` | Cualquier carácter excepto salto de línea | `c.t` | `cat`, `cot`, `c9t` |
| `\.` | Punto literal (escapado) | `www\.example` | `www.example` |
| `\d` | Un dígito (0-9) | `\d` | `0`, `5`, `9` |
| `\w` | Letra, dígito o `_` | `\w` | `a`, `1`, `_` |
| `\s` | Espacio en blanco | `\s` | `" "`, `"\t"` |
| `\D` | Todo lo que NO es dígito | `\D` | `a`, `#`, `!` |
| `\W` | Todo lo que NO es `\w` | `\W` | `!`, `@`, `" "` |
| `\S` | Todo lo que NO es espacio | `\S` | `a`, `1`, `#` |

### Cuantificadores

| Patrón | Significa | Ejemplo | Encuentra |
|--------|-----------|---------|-----------|
| `{n}` | Exactamente n veces | `\d{3}` | `123` |
| `{n,m}` | Entre n y m veces | `\d{2,4}` | `12`, `123`, `1234` |
| `+` | Una o más veces | `\d+` | `45`, `7890` |
| `*` | Cero o más veces | `\d*` | `""`, `1`, `123` |
| `?` | Cero o una vez (opcional) | `\d?` | `""`, `5` |

### Conjuntos y grupos

| Patrón | Significa | Ejemplo | Encuentra |
|--------|-----------|---------|-----------|
| `[abc]` | Cualquiera de estos caracteres | `[aeiou]` | `a`, `e`, `i` |
| `[A-Z]` | Rango de mayúsculas | `[A-Z]+` | `HOLA`, `RUC` |
| `[a-z]` | Rango de minúsculas | `[a-z]+` | `hola`, `mundo` |
| `[A-Za-z]+` | Solo letras | `[A-Za-z]+` | `Hola`, `mundo` |
| `[^abc]` | Todo excepto estos caracteres | `[^0-9]` | `a`, `!`, `#` |
| `(abc)` | Grupo de captura | `(\d{2})/(\d{2})` | captura `01` y `12` |

### Anclas y límites

| Patrón | Significa | Ejemplo | Encuentra |
|--------|-----------|---------|-----------|
| `^` | Inicio de la cadena | `^\d+` | línea que empieza con dígito |
| `$` | Fin de la cadena | `\d+$` | línea que termina con dígito |
| `\b` | Límite de palabra | `\bRUC\b` | `RUC` solo, no `RUCA` |

### Funciones principales

| Función | Significa | Devuelve |
|---------|-----------|----------|
| `re.search(p, txt)` | Primera coincidencia en cualquier parte | Objeto Match o `None` |
| `re.match(p, txt)` | Coincidencia solo al inicio | Objeto Match o `None` |
| `re.findall(p, txt)` | Todas las coincidencias | Lista |
| `re.sub(p, repl, txt)` | Reemplaza coincidencias | String nuevo |
| `re.split(p, txt)` | Divide el texto por el patrón | Lista |

- `search()` te devulve un objeto si encuentra si no te devuelve un None.
- `group()` extrae el texto encontrado. 



```python

text = "Factura E001-5834 RUC: 20610320351 Total: S/ 20.00"

# Buscar 11 digitos seguidos
print(re.search(r'\d{11}', text))
print(re.search(r'\d{20}', text)) 
result = re.search(r'\d{11}', text)
print(result.group())
```

```python
# Obtener la serie y el número
print(re.search(r'[A-Z]\d{3}-\d{4}', text))
resultado = re.search(r'[A-Z]\d{3}-\d{4}', text)
print(resultado.group())
```

```python
# Ahora obtener el precio con los 2 decimales
print(text)
print(re.search(r'\d{2}\.\d{2}', text))
total = re.search(r'\d{2}\.\d{2}', text)
print(total.group())
```
