# Tipos de datos

Cada objeto en memoria tiene su propio tipo de dato definido desde que se crea.

## Tipo de dato: Texto

Es un objeto de la clase `str` que representa una secuencia ordenada
de caracteres, donde cada carácter tiene una posición llamada índice.
Es **inmutable**, lo que significa que una vez creado el objeto en memoria,
no puede ser modificado.

````python
nombre_curso = "Python"

print(nombre_curso[0])
print(nombre_curso[1])
print(nombre_curso[2])
print(nombre_curso[3])
print(nombre_curso[4])
print(nombre_curso[5])
````

### Como funciona los índices en la cádena



````python
frase = 'Hola'
print(frase[0])    # H → primer carácter desde la izquierda
print(frase[0:4])  # Hola → desde posición 0 hasta antes de la 4
print(frase[-4])   # H → primer carácter desde la derecha
print(frase[:])    # Hola → toda la cadena
````

### Métodos

Al ser un objeto de la clase `str`, tiene sus propios métodos incluidos:

- `.upper()` → convierte el texto a mayúsculas
- `.lower()` → convierte el texto a minúsculas
- `.strip()` → elimina espacios al inicio y al final
- `.replace()` → reemplaza un texto por otro

````python
nombre_curso = "Rust"
print(nombre_curso)
nombre_curso.upper() ## Se crea en memoria pero si no se hace referencia se pierde
nombre_curso = nombre_curso.upper()
print(nombre_curso)
````

## Tipo de dato: Números

````python
edad = 25
print(type(edad))
````

**números complejos**: Un número complejo es la combinación de una parte real y una parte imaginaria.

**float**: Números con decimales. Esto siguen el estándar de IEEE 745 que tiene una precisión limitada de 15 a 17 dígitos.

````python
from decimal import Decimal

precio = 3.14
print(type(precio))

## Precisión limitada

print(0.1+0.2) ## algunos dedimales no tiene representación exacta en binario

## Utilizando un paquete para precisión exacta deben estar en comillas

subtotal = Decimal("0.1")+Decimal("0.2")
print(subtotal)
print(type(subtotal))
````

### Separadores numéricos con `_`

Se usa el guion bajo (`_`) para separar visualmente números largos
y mejorar la legibilidad del código. Python ignora los `_` internamente,
el valor almacenado en memoria es el mismo.

````python
poblacion = 1_000_000      # igual a 1000000
precio = 1_234_567.89      # igual a 1234567.89

print(poblacion)
print(type(poblacion))
````

### Multilpe Assignment
Asigna valores a varias variables en una sola línea. Python lee primero
el lado derecho, crea los objetos en memoria y luego asigna cada etiqueta
a su valor correspondiente por posición. Internamente usa tuplas para
empaquetar y desempaquetar los valores.

**Sintaxis:**
```python
a, b = valor_a, valor_b

````python
a, b = 10, 20

print(f'El valor de a es {a} y b es {b}')
````

````python
# Asignar el valor a varias variables a la vez
a = b = c = 2
print(a)
print(b)
print(c)
````

````python
# intercambiar valores
a = 200
b = 100

a,b = b,a
print(a)
print(b)
````

````python
# Tupla como se ve

tupla_valores = 10, 20, 30
print(type(tupla_valores))
````

### Listas

Es un tipo de dato compuesto que agrupa múltiples valores bajo una misma
etiqueta. Es **mutable**, lo que significa que sus elementos pueden
modificarse, agregarse o eliminarse sin crear un nuevo objeto en memoria.

````python
# ✅ Lista homogénea → recomendado
list_num = [1, 2, 3]
list_text = ["Miguel", "Pedro", "Luis"]
# ⚠️ Lista mixta → no recomendado mejor tupla
list_mix = [1, "Miguel", True]

print(type(list_num))
print(list_num)
print(list_text)
print(list_mix)
````

````python
list_text[0]
````

Vemos como funciona internamente una lista en memoria:
 - **append** → agrega un solo elemento al final de la lista
 - **extend** → agrega los elementos de otra lista uno por uno al final



````python
estudiantes = ["Juan", "Pedro", "Miguel"]
print(estudiantes)
print(f'Referencia principal: {id(estudiantes)}')
print(f'Referencia del elemento 0: {id(estudiantes[0])}')
print(f'Referencia del elemento 1: {id(estudiantes[1])}')
print(f'Referencia del elemento 2: {id(estudiantes[2])}')

# Cambiamos el elmento 0 por otro
estudiantes[0] = "Anita"
print(estudiantes)
print(f'Referencia principal: {id(estudiantes)}')
print(f'Referencia del elemento 0: {id(estudiantes[0])}')

# Agregar el elemento

estudiantes.append("Alejandro")
print(estudiantes)

estudiantes.extend(["Alexis","Mauro"])
print(estudiantes)
````

### Tupla

Es un tipo de dato compuesto inmutable y ordenada que agrupa múltiples valores en ( ) bajo una misma etiqueta. Al ser inmutable, una vez creada en memoria no puede modificarse.

````python
# Crear una tupla para estudiantes quiero guardar el id_estudiante, nombre, edad
tupla_estudiantes = (1, "Juan", 30)
print(tupla_estudiantes)
print(type(tupla_estudiantes))
````

### Diccionario

Es un tipo de dato compuesto mutable que almacena pares de clave-valor. A diferencia de la lista que usa índices numéricos, el diccionario usa claves para acceder a cada valor.

- El método `.get()` permite especificar un valor por defecto en caso de que no exista esa clave y no tener un error.
- El método `.pot()` elimina elementos.
- El método `.keys()` va a mostrar solamente las claves.
- El método `.values()` va a mostrar solamente los valores.

````python
personas = {
    "nombre":"Miguel",
    "edad":32, 
    #"ciudad":"Lima"
}

# Accerder a cada valor
print(personas["nombre"])
print(personas["edad"])
print(personas.get("ciudad", "No existe"))

# Agregar o modificar
personas["email"] = "mg@gmail.com"
personas["edad"] = 45

print(personas["email"])
print(personas["edad"])

# Eliminar
personas.pop("email")

# ver claves o valores
print(personas.keys())
print(personas.values())
print(personas.items())

# Recorrer

for clave, valor in personas.items():
    print(f'{clave}: {valor}')
````

- **Slice** no es función ni método, es una sintaxis propia de Python que funciona en cualquier secuencia:
- Regla clave: el fin nunca se incluye por eso se le aumenta uno, igual que range().
- sintaxis: `secuencia[inicio:fin:paso]` cada parte es opcional

````python
message = "Hola Mundo"
print(message[0:4:1])
print(message[:4])
````

````python
numeros = [0, 1, 2, 3, 4, 5]
#índices:   0  1  2  3  4  5

print(numeros[1:4])    # → [1, 2, 3]      desde índice 1 hasta 4 (sin incluir)
print(numeros[::2])   # → [0, 2, 4]      de 2 en 2
print(numeros[::-1])   # → [5, 4, 3, 2, 1, 0]  invertido
print(numeros[2:])   # → [2, 3, 4, 5]   desde índice 2 hasta el final
print(numeros[:3])   # → [0, 1, 2]      desde el inicio hasta índice 3 (sin incluir)
````
