## ¿Qué es una variable?

No es una caja ni un contenedor, es una referencia a un objeto en memoria.



Cuando trabajamos con enteros, strings, float, booleano no se modifica el objeto ya creado, si no se va a crear una nueva referencia.



**Inmutabilidad:**  Significa que una vez creado el objeto en memoria, no se puede modificar. Si "cambias" el valor, Python crea un objeto nuevo.

````python
edad = 42
print("referencia 1: ", id(edad))
edad = 12
print("referencia 2: ", id(edad))
x = edad
print("referencia 3: ", id(x))
edad
````

**Mutables:** Significa que el objeto en memoria sí se puede modificar sin crear uno nuevo

````python
lista = [1,2,3]
print("referencia 1:",id(lista))
print(lista)

lista.append(4)
print("referencia 1:",id(lista))
print(lista)
````

### Convención de nomenclatura

**snake_case** es el estilo de escritura estándar en Python para nombrar variables y funciones.

✅ Se escribe todo en minúsculas.  
✅ Las palabras se separan con guion bajo (`_`). Ejemplo: `id_empleado`  
✅ **PEP 8** es la guía de estilo oficial de Python que define cómo debe escribirse el código.

### Constantes

Es una convención que comunica que este valor no debe cambiar.
En Python no existe una constante real como en otros lenguajes,
es solo un acuerdo entre programadores.
Se escribe en **MAYÚSCULAS** con guion bajo para separar palabras:
```python
MAX_INTENTOS = 3
PI = 3.14159
NOMBRE_EMPRESA = "Anthropic"
```
