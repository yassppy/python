# Estructuras de control

Controlan el flujo de ejecución del programa

## Condicionales

### IF

Es utilizada para tomar decisiones, mediante una condición si esto es verdad vas a realizar esto, si no, realiza lo siguiente.

```python
# Problema de clasificador de números si es mayor a 0, "positvo", menor "negativo" y si es 0 "cero"
def classify(num):
    if num == 0:
        return "cero"
    elif num > 0:
        return "positivo"
    else:
        return "negativo"

print(classify(10))
print(classify(0))
print(classify(-10))
```

### match / case

Es el switch pero con otro nombre, solo evalúa un valor y elige un camino.

```python
lenguaje = "python"
match lenguaje:
    case "python":
        print("Snake! 🐍")
    case "javascript":
        print("Browser! 🌐")
    case _:
        print("Otro lenguaje")
```

## Loop(Bucles)

Es una estructura de control que permite ejecutar el mismo bloque de código varias veces sin necesidad de escribirlo repetidamente. 

### For Loop

El for se utiliza cuando queremos recorrer cualquier secuencia iterable, va elemento por elemento hasta que se acabe. Cada vuelta se llama iteración y la variable temporal guarda el elemento actual. Se utiliza cuando queremos aplicar un bloque de código de manera finita (tiene un inicio y un final). 

1. Ejemplo con String

```python
for letra in "hola":
	print(letra)
```

2. Ejemplo con lista

```python
for frutas in ["manzana", "Pera"]:
	print(frutas)
```

3. Ejemplo con tupla

```python
for edad in (11, 12, 17):
	print(edad)
```

4. Ejemplo con diccionario

```python
for clave, valor in {"nombre":"Juan", "edad":20}.items():
	print(f'{clave}: {valor}')
```

```python
# Ejercicio Escribe una función sumar(numeros) que reciba una lista de números y retorne la suma total sin usar sum().

def sum_num(numeros):
	total = 0
	for i in numeros:
		total = i + total
	return total

numeros = [1, 2, 3, 4, 5]
print(sum_num(numeros))
```

- range() es una función built-in (incorporada en Python), genera un objeto de tipo range:

- El número final nunca se incluye, siempre se queda fuera. Por eso siempre se le aumenta uno más.
- `range(inicio, fin, paso)`

```python
for a in range (5):
	print(a)
print()
for i in range (1, 5, 1):
	print(i)

print(f"\n {type(range(1))}")
```

- Comprehensions con for
- Es una forma corta y elegante de crear una lista usando un for en una sola línea.
- Sintaxis: `[expresión for variable in secuencia]`

```python
num = [i * 2 for i in range(5)]
print(num)

pares = [i for i in range(10) if i % 2 == 0]
print(pares)
```

### while loop

Se ejecuta mientras una condición sea verdadera. A diferencia del for, no recorre una secuencia, sino que repite hasta que algo cambie. Se aplica cuando queremos utilizar repeticiones y no sabemos cuándo finaliza. ⚠️ Si la condición nunca se vuelve False, el programa se congela en un loop infinito y debes detenerlo manualmente con Ctrl + C.

```python
def cuenta_regresiva(num):
	contador = num
	while contador>0:#condición verdadera hasta que sea falso
		print(contador)
		contador-=1

print(cuenta_regresiva(5))

```
